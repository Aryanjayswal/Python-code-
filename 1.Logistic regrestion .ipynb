# Logistic Regression

Logistic Regression is a classification algorithm used to predict whether something belongs to one category or another (e.g., spam or not spam, pass or fail, disease or no disease).

Even though it has "regression" in its name, Logistic Regression is actually used for classification, not regression!

#### ðŸ“Œ How Does Logistic Regression Work?
It takes input features (X) and applies a mathematical function to produce an output between 0 and 1.
The output represents the probability of belonging to a particular class.
We use a threshold (usually 0.5):
* If probability â‰¥ 0.5, classify as 1 (positive class).
* If probability < 0.5, classify as 0 (negative class).

##### âš¡ The Key Formula: Sigmoid Function
* Unlike Linear Regression, which predicts any number, Logistic Regression uses the sigmoid function to squeeze outputs between 0 and 1:




#### ðŸ”¥ Why Not Use Linear Regression Instead?
* Linear Regression predicts values outside 0 and 1, which is not useful for classification.
* Logistic Regression keeps outputs in range and converts them into probabilities.

* 1ï¸âƒ£ Binary Logistic Regression â†’ Two categories (e.g., spam or not spam).
* 2ï¸âƒ£ Multinomial Logistic Regression â†’ More than two categories (e.g., predicting red, blue, or green).
* 3ï¸âƒ£ Ordinal Logistic Regression â†’ Ordered categories (e.g., low, medium, high satisfaction).
